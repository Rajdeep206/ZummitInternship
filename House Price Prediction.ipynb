```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
house_data=pd.read_csv('House price prediction data.csv')
house_data.head()
```

```python
house_data.describe()
```

```python
house_data_latest=house_data.set_index('price')
house_data_latest.head()
```

```python
house_data_latest.info()
```

```python
house_data_latest=house_data_latest.drop(0.0)
house_data_latest=house_data_latest.drop(['date','street','city','statezip','country'],axis=1)
```

```python
house_data_latest.info()
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
from sklearn.model_selection import train_test_split
house_data_latest=house_data_latest.reset_index()
x=house_data_latest.drop('price',axis=1)
y=house_data_latest['price']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train_scaled=sc.fit_transform(x_train)
x_test_scaled=sc.transform(x_test)
```

```python
model.fit(x_train_scaled,y_train)
```

```python
model.score(x_test_scaled,y_test)
```
